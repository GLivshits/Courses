```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Задание функции плотности

```python
cc = 4
gamma_rv = sts.gamma(a = cc)
sample = gamma_rv.rvs(1000)
```

# Теоретическая функция плотности и соответсвующая гистограмма выборки

```python
x = np.linspace(min(sample),max(sample),1000)
gamma_pdf = gamma_rv.pdf(x)
plt.plot(x, gamma_pdf, label = "theoretical DF")
plt.hist(sample, density=True, label = "histogram")
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

# Проверка ЦПТ при n = 5

```python
data = np.zeros((1000))
n = 5
xn = np.zeros((n))
for i in range (0,1000):
    xn = gamma_rv.rvs(n)
    data[i] = np.mean(xn)
norm_rv = sts.norm(loc = cc, scale = (cc/n)**0.5)
x = np.linspace(min(data),max(data),1000)
norm_pdf = norm_rv.pdf(x)
plt.plot(x, norm_pdf, label = "theoretical DF")
plt.hist(data, density = True, label = "histogram")
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```
