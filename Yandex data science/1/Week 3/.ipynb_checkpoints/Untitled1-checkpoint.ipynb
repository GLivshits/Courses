```python
import numpy as np
from scipy import linalg
import math
from matplotlib import pyplot as plt
```

```python
def f(x):
    return math.sin(x/5)*math.exp(x/10) + 5*math.exp(-x/2)
```

```python
x2 = np.array([[1,1],[1,15]])
x3 = np.array([[1,1,1],[1,8,8**2],[1,15,15**2]])
x4 = np.array([[1,1,1,1],[1,4,4**2,4**3],[1,10,100,1000],[1,15,15**2,15**3]])
```

```python
f2 = np.array([f(x2[0,1]),f(x2[1,1])])
f3 = np.array([f(x3[0,1]),f(x3[1,1]),f(x3[2,1])])
f4 = np.array([f(x4[0,1]),f(x4[1,1]),f(x4[2,1]),f(x4[3,1])])
```

```python
a2 = np.array(linalg.solve(x2,f2))
a3 = np.array(linalg.solve(x3,f3))
a4 = np.array(linalg.solve(x4,f4))
```

```python
xx = []
for i in range (0,150):
    xx.append(0.1*i)
ff = []
for i in range (0,150):
    ff.append(f(xx[i]))
```

```python
plt.plot(xx,ff)
```

```python
ff2 = []
ff3 = []
ff4 = []
for i in range(0,150):
    ff2.append(a2[0]+a2[1]*xx[i])
    ff3.append(a3[0]+a3[1]*xx[i]+a3[2]*xx[i]**2)
    ff4.append(a4[0]+a4[1]*xx[i]+a4[2]*xx[i]**2+a4[3]*xx[i]**3)
```

```python
plt.plot(xx,ff2)
```

```python
plt.plot(xx,ff3)
```

```python
plt.plot(xx,ff4)
```
