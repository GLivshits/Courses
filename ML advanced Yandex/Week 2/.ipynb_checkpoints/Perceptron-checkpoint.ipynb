```python
import pandas as pd
import numpy as np
from sklearn import *
from matplotlib import pyplot as plt
import math
```

```python
train = pd.read_csv('perceptron-train.csv', engine = 'python', header = None)
test = pd.read_csv('perceptron-test.csv', engine = 'python', header = None)
```

```python
X_train = train.drop([0], axis = 1)
y_train = train[0]
X_test = test.drop([0], axis = 1)
y_test = test[0]
X = pd.concat([X_train, X_test])
```

```python
alls = preprocessing.StandardScaler().fit_transform(X)
```

```python
X_train_scaled = preprocessing.StandardScaler().fit_transform(X_train)
X_test_scaled = preprocessing.StandardScaler().fit_transform(X_test)
```

```python
model = linear_model.Perceptron()
```

```python
model = linear_model.Perceptron(random_state = 241, max_iter = 5)
model.fit(X_train, y_train)
accuracy = metrics.accuracy_score(y_test, model.predict(X_test))
model = linear_model.Perceptron(random_state = 241, max_iter = 5)
model.fit(X_train_scaled, y_train_scaled)
accuracy_scaled = metrics.accuracy_score(y_test, model.predict(X_test_scaled))
```

```python
train_scaled = preprocessing.StandardScaler().fit_transform(train)
test_scaled = preprocessing.StandardScaler().fit_transform(test)
```

```python
X_train_scaled = np.transpose(train_scaled[:, 1:])
y_train_scaled = np.transpose(train_scaled[:, 0].flatten())
X_test_scaled = np.transpose(test_scaled[:, 1:])
y_test_scaled = np.transpose(test_scaled[:, 0].flatten())
```

```python
file = open('answer_2_1.txt', 'w')
file.write('{}'.format(round(accuracy_scaled - accuracy, 3)))
file.close()
```
