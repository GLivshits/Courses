```python
df_6 = df['Name'][df['Sex'] == 'female']
x = []
for item in df_6:
    x.append(item.split(','))
```

```python
x_1 = []
for item in x:
    x_1.append(item[1])
```

```python
x_2 = []

for item in x_1:
    
    print(item)
    if 'Mrs' in item:
        
        item = item.split('. ')[1]
        
        if '(' in item:
            
            item = item.split('(')[1]
            item = item.split(' ')[0]
        
        else:
        
            item = item.split(' ')[0]
        
        item = item.replace(')', '')
        
        x_2.append(item)
        
    if 'Miss' in item:
        
        item = item.split('. ')[1]
        item = item.split(' ')[0]
        x_2.append(item)
        
```
